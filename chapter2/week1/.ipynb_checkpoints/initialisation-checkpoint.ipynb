```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import sklearn.datasets
from init_utils import sigmoid, relu, compute_loss, forward_propagation, backward_propagation
from init_utils import update_parameters, predict, load_dataset, plot_decision_boundary, predict_dec

%matplotlib inline
plt.rcParams['figure.figsize'] = (7.0, 4.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# load image dataset: blue/red dots in circles
train_X, train_Y, test_X, test_Y = load_dataset()
train_X.shape
```

```python
def initial_parameters(layer_dims):

    np.random.seed(3) 

    init_params = []
    for i in range(1, len(layer_dims)):
        w1 = np.zeros((layer_dims[i], layer_dims[i-1]))
        b1 = np.zeros((layer_dims[i], 1))
        init_params.append((w1, b1))

    return init_params
    
```

```python
def forward_pass(X, initial_params):
    
    #Parameters which are Z and A of a given layer
    parameters = []
    cache = []

    #Let's pass for the next layers
    inp = X
    # print(X.shape)
    for i in range(len(initial_params) - 1):
        Z = np.dot(initial_params[i][0],inp) + initial_params[i][1]
        A = relu(Z)
        parameters.append(Z)
        parameters.append(A)
        cache.append(initial_params[i][0])
        cache.append(initial_params[i][1])
        # print("The shape of w1 is: ", initial_params[i][0].shape)
        # print("The shape of b1 is: ", initial_params[i][1].shape)
        inp = A
        # print("The shape of A is: ", A.shape)

    Zlast = np.dot(initial_params[-1][0], A) + initial_params[-1][1]
    yest = sigmoid(Zlast)
    parameters.append(Zlast)
    parameters.append(yest)
    cache.append(initial_params[-1][0])
    cache.append(initial_params[-1][1])
    # print("The shape of w1 is: ", initial_params[-1][0].shape)
    # print("The shape of b1 is: ", initial_params[-1][1].shape)
    # print("The shape of A is: ", yest.shape)
    
    return parameters, yest, cache
    
```

```python
def compute_loss(a3, Y):

    m = Y.shape[1]
    # print("The shape of Y is: ",Y.shape)
    logprobs = np.multiply(-np.log(a3),Y) + np.multiply(-np.log(1 - a3), 1 - Y)
    # print("The shape of logprobs: ", logprobs.shape)
    loss = np.nansum(logprobs)/m

    return loss
    
```

```python
def back_pass(X,Y,parameters,cache):
    #Write the code for back propagation
    #Need to start from the end

    (w1, b1, w2, b2, w3, b3) = cache
    (z1, a1, z2, a2, z3, a3) = parameters

    m = X.shape[1]

    dz3 = 1./m * (a3 - Y)
    dW3 = np.dot(dz3, a2.T)
    db3 = np.sum(dz3, axis=1, keepdims = True)

    da2 = np.dot(w3.T, dz3)
    dz2 = np.multiply(da2, np.int64(a2 > 0))
    dW2 = np.dot(dz2, a1.T)
    db2 = np.sum(dz2, axis=1, keepdims = True)

    da1 = np.dot(w2.T, dz2)
    dz1 = np.multiply(da1, np.int64(a1 > 0))
    dW1 = np.dot(dz1, X.T)
    db1 = np.sum(dz1, axis=1, keepdims = True)

    gradients = {"dz3": dz3, "dW3": dW3, "db3": db3,
                 "da2": da2, "dz2": dz2, "dW2": dW2, "db2": db2,
                 "da1": da1, "dz1": dz1, "dW1": dW1, "db1": db1}
    
    return gradients
```

```python
def update_params(cache, gradients, learning_rate):

    (w1, b1, w2, b2, w3, b3) = cache

    w1 += -(learning_rate) * gradients['dW1']
    b1 += -(learning_rate) * gradients['db1']
    w2 += -(learning_rate) * gradients['dW2']
    b2 += -(learning_rate) * gradients['db2']
    w3 += -(learning_rate) * gradients['dW3']
    b3 += -(learning_rate) * gradients['db3']

    p = []
    p.append((w1, b1))
    p.append((w2, b2))
    p.append((w3, b3))

    return p
```

```python

#Creating a 3 layered model
def model(X, Y, learning_rate = 0.01, num_iterations = 15000, print_cost = True, initialization = "he"):

    """
    Implements a three-layer neural network: LINEAR->RELU->LINEAR->RELU->LINEAR->SIGMOID.
    
    Arguments:
    X -- input data, of shape (2, number of examples)
    Y -- true "label" vector (containing 0 for red dots; 1 for blue dots), of shape (1, number of examples)
    learning_rate -- learning rate for gradient descent 
    num_iterations -- number of iterations to run gradient descent
    print_cost -- if True, print the cost every 1000 iterations
    initialization -- flag to choose which initialization to use ("zeros","random" or "he")
    
    Returns:
    parameters -- parameters learnt by the model
    """

    layer_dims = [X.shape[0], 10, 10, 1]

    init_params = initial_parameters(layer_dims)
    params = init_params
    costs = []

    #Calculating the loss during gradient descent
    for _ in range(num_iterations):
        
        #Forward pass
        parameters, Yest, cache = forward_pass(X, params)
        
        #Compute the loss 
        cost = compute_loss(Yest, Y)
        
        #Backward pass
        grads = back_pass(X, Y, parameters, cache)
        # print(grads)
        
        #Update parameters
        params = update_params(cache, grads, learning_rate)

        if print_cost and _ % 1000 == 0:
            print("Cost after iteration {}: {}".format(_, cost))
            costs.append(cost)

    plt.plot(costs)
    plt.ylabel('cost')
    plt.xlabel('iterations (per hundreds)')
    plt.title("Learning rate =" + str(learning_rate))
    plt.show()

    return params
    
```

```python
def predict(X, y, parameters):
    """
    This function is used to predict the results of a  n-layer neural network.
    
    Arguments:
    X -- data set of examples you would like to label
    parameters -- parameters of the trained model
    
    Returns:
    p -- predictions for the given dataset X
    """
    
    m = X.shape[1]
    p = np.zeros((1,m), dtype = np.int32)
    
    # Forward propagation
    parame, a3, caches = forward_pass(X, parameters)

    predictions = (a3>0.5)
    
    # convert probas to 0/1 predictions
    for i in range(0, a3.shape[1]):
        if a3[0,i] > 0.5:
            p[0,i] = 1
        else:
            p[0,i] = 0

    # print results
    print("Accuracy: "  + str(np.mean((p[0,:] == y[0,:]))))
    
    return predictions, p
```

```python
def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    print(np.c_[xx.ravel(), yy.ravel()])
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    print("The shape of the model before reshaping is: ", Z.shape)
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=y, cmap=plt.cm.Spectral)
    plt.show()
    return Z.shape
```
